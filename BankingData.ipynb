```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv('E:\\Datasets\\classification\\bnk\\BankData.csv',sep=';')
```

```python
data.head()
```

```python
data.info()
```

```python
data['y'].value_counts()
```

```python
feature=data.iloc[:,0:20]
target=data['y']
```

```python
feature.head()
```

```python
Numerical_data=feature.select_dtypes(exclude='object')
Numerical_data.head()
```

```python
categorical_data=feature.select_dtypes(include='object')
```

```python
categorical_encoding=pd.get_dummies(categorical_data[categorical_data.columns])
```

```python
categorical_encoding.head()
```

```python
merge_data=[Numerical_data,categorical_encoding]
training_data=pd.concat(merge_data,axis=1)
```

```python
training_data.head()
```

```python
## feature Selection process
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectFromModel
```

## ImbalancedData handling

```python
from imblearn.over_sampling import SMOTE
smot=SMOTE(random_state=0, n_jobs=8)
```

```python
x,y=smot.fit_resample(training_data,target)
```

```python
## scalling 
from sklearn.preprocessing import MinMaxScaler
scld=MinMaxScaler()
scld.fit(x)
scaledTran=scld.transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(scaledTran,y,test_size=0.3,random_state=0)
```

```python
## Model gradinet Boosting classifier
from sklearn.ensemble import GradientBoostingClassifier
Gradient_model=GradientBoostingClassifier(n_estimators=100, learning_rate=1.0,max_depth=1, random_state=0)
Gradient_model.fit(x_train,y_train)
```

```python
gboost_predict=Gradient_model.predict(x_test)
```

```python
from sklearn import metrics
print('classification report',metrics.classification_report(y_test,gboost_predict))
```

```python
print('accuracy',metrics.accuracy_score(y_test,gboost_predict))
```

```python
from xgboost import XGBClassifier
```

```python
xgb = XGBClassifier(n_estimators=100, max_depth=5,min_child_weight= 1,learning_rate=0.1)
xgb.fit(x_train,y_train)
```

```python
xgBoost_pred=xgb.predict(x_test)
print('classification report',metrics.classification_report(y_test,xgBoost_pred))
```

```python
print('accuracy',metrics.accuracy_score(y_test,xgBoost_pred))
```

## hyper tuning XGBoost

```python
from sklearn.model_selection import GridSearchCV
param_test1 = {
 'max_depth':range(3,10,2),
 'min_child_weight':range(1,6,2)
}
gsearch1 = GridSearchCV(estimator = XGBClassifier( learning_rate =0.1, n_estimators=140, max_depth=5,
 min_child_weight=1, gamma=0, subsample=0.8, colsample_bytree=0.8,
 objective= 'binary:logistic', nthread=4, scale_pos_weight=1, seed=27), param_grid = param_test1, scoring='roc_auc',n_jobs=4, cv=5)
gsearch1.fit(x_train,y_train)
```

```python
gsearch1.best_params_, gsearch1.best_score_
```

```python
predgxgb=gsearch1.predict(x_test)
```

```python
print('classification report',metrics.classification_report(y_test,predgxgb))
```

## tensorflow 

```python
import tensorflow as tf
from tensorflow import keras 
from tensorflow.keras import layers,Model,Sequential
```

```python
model=Sequential()
model.add(layers.Dense(32,activation='relu',input_shape=(63,)))
model.add(layers.Dense(64,activation='relu'))
model.add(layers.Dense(1,activation='sigmoid'))
model.compile(
    optimizer="rmsprop",
    loss=tf.keras.losses.BinaryCrossentropy(),
    metrics=["accuracy"]
)
```

```python
model.fit(x_train,y_train,batch_size=10,epochs=10,verbose=2)
```
